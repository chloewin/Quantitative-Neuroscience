```python
import pandas as pd
```

```python
clinical_data = pd.read_csv("./../data/ADNI_clinical_characteristics.csv")

clinical_data.head()
```

```python
clinical_data["BMI"] = clinical_data.VSWEIGHT / (clinical_data.VSHEIGHT ** 2)

# height units 1 for in 2 for cm
# weight units 1 for lb, 2 for kg
lbs = clinical_data.VSWEIGHT
lbs[clinical_data.VSWTUNIT == 2] = clinical_data.VSWEIGHT[clinical_data.VSWTUNIT == 2] * 2.20462

ins = clinical_data.VSHEIGHT
ins[clinical_data.VSHTUNIT == 2] = clinical_data.VSHEIGHT[clinical_data.VSHTUNIT == 2] * 0.393701

clinical_data["BMI"] = 703 * lbs / (ins ** 2)
clinical_data.head()
print(clinical_data["BMI"].isna().sum())
```
